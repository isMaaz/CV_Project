```python
import subprocess, sys
subprocess.check_call([sys.executable, "-m", "pip", "install",
                       "nibabel", "scikit-image", "scipy", "matplotlib",
                       "kagglehub", "-q"])

import os, gc, csv, warnings
import numpy as np
import nibabel as nib
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
from pathlib import Path
from scipy.ndimage import gaussian_filter, uniform_filter, median_filter
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim
from tqdm import tqdm
import kagglehub

warnings.filterwarnings("ignore")

MAX_VOLUMES    = 100
MAX_SLICES     = 20
GAUSSIAN_SIGMA = 1.0
MEAN_SIZE      = 3
MEDIAN_SIZE    = 3
SAVE_EVERY_N   = 5
OUT_DIR        = Path("output")
FILTER_NAMES   = ["gaussian", "mean", "median"]

for d in [OUT_DIR / "before",
          OUT_DIR / "after" / "gaussian",
          OUT_DIR / "after" / "mean",
          OUT_DIR / "after" / "median",
          OUT_DIR / "comparisons"]:
    d.mkdir(parents=True, exist_ok=True)

print("Downloading dataset via kagglehub...")
path = kagglehub.dataset_download("awansaad6797/cancer-dataset")
print(f"Path to dataset files: {path}")

print("\nScanning dataset folder...")
all_files = sorted(Path(path).rglob("*"))
extensions = {}
for f in all_files:
    if f.is_file():
        ext = "".join(f.suffixes).lower()
        extensions[ext] = extensions.get(ext, 0) + 1
print("File types found:")
for ext, count in sorted(extensions.items()):
    print(f"  {ext if ext else '(no extension)'}: {count} files")

print("\nSample paths (up to 10):")
for f in [f for f in all_files if f.is_file()][:10]:
    print(f"  {f.relative_to(path)}")

all_nii = sorted(f for f in Path(path).rglob("*.nii.gz") if f.is_file())
if len(all_nii) == 0:
    all_nii = sorted(f for f in Path(path).rglob("*.nii") if f.is_file())
    print(f"\nNo .nii.gz found â€” using .nii: {len(all_nii)} actual files")
if len(all_nii) == 0:
    all_nii = sorted(f for f in Path(path).rglob("*.mha") if f.is_file())
    print(f"No .nii found â€” using .mha: {len(all_nii)} files")
if len(all_nii) == 0:
    raise RuntimeError(
        f"No supported MRI files found (.nii.gz / .nii / .mha).\n"
        f"Dataset path: {path}"
    )

all_nii = [f for f in all_nii if "seg" not in f.name.lower()]
print(f"\nTotal usable MRI files (non-seg): {len(all_nii)}")
all_nii = all_nii[:MAX_VOLUMES]
print(f"Using {len(all_nii)} volumes (MAX_VOLUMES={MAX_VOLUMES})")

def normalise(slc):
    mn, mx = slc.min(), slc.max()
    if mx > mn:
        return ((slc - mn) / (mx - mn + 1e-8)).astype(np.float32)
    return np.zeros_like(slc, dtype=np.float32)

def apply_filters(slc):
    return {
        "gaussian": gaussian_filter(slc, sigma=GAUSSIAN_SIGMA),
        "mean":     uniform_filter(slc,  size=MEAN_SIZE),
        "median":   median_filter(slc,   size=MEDIAN_SIZE),
    }

def compute_metrics(orig, filt):
    p = psnr(orig, filt, data_range=1.0)
    s = ssim(orig, filt, data_range=1.0)
    p = 100.0 if np.isinf(p) else float(p)
    return round(p, 4), round(float(s), 4)

def save_png(arr, fpath):
    fig, ax = plt.subplots(figsize=(3, 3), dpi=80)
    ax.imshow(arr, cmap="gray", vmin=0, vmax=1)
    ax.axis("off")
    fig.savefig(fpath, bbox_inches="tight", pad_inches=0)
    plt.close(fig)

def save_comparison(orig, filtered, mets, fpath):
    titles = ["Original"] + [k.capitalize() for k in filtered]
    images = [orig] + list(filtered.values())
    fig, axes = plt.subplots(1, 4, figsize=(16, 4))
    for ax, title, img in zip(axes, titles, images):
        ax.imshow(img, cmap="gray", vmin=0, vmax=1)
        ax.axis("off")
        if title == "Original":
            ax.set_title(title, fontsize=9)
        else:
            p_val, s_val = mets[title.lower()]
            ax.set_title(f"{title}\nPSNR={p_val:.2f} SSIM={s_val:.4f}",
                         fontsize=8)
    fig.tight_layout()
    fig.savefig(fpath, dpi=90, bbox_inches="tight")
    plt.close(fig)

def centre_slice_indices(n_slices, max_slices):
    if n_slices <= max_slices:
        return list(range(n_slices))
    start = (n_slices - max_slices) // 2
    return list(range(start, start + max_slices))

def load_volume(nii_path):
    suffix = "".join(nii_path.suffixes).lower()
    if suffix in [".nii.gz", ".nii"]:
        img = nib.load(str(nii_path))
        return img.get_fdata(dtype=np.float32)
    elif suffix in [".mha", ".mhd"]:
        try:
            import SimpleITK as sitk
        except ImportError:
            subprocess.check_call([sys.executable, "-m", "pip",
                                   "install", "SimpleITK", "-q"])
            import SimpleITK as sitk
        img = sitk.ReadImage(str(nii_path))
        vol = sitk.GetArrayFromImage(img).astype(np.float32)
        return np.transpose(vol, (1, 2, 0))
    else:
        raise ValueError(f"Unsupported format: {suffix}")

csv_path = OUT_DIR / "metrics.csv"
csv_rows = []
volumes_processed = 0

with open(csv_path, "w", newline="") as csv_file:
    writer = csv.DictWriter(
        csv_file,
        fieldnames=["subject", "slice", "filter", "psnr", "ssim"]
    )
    writer.writeheader()

    for nii_path in tqdm(all_nii, desc="Volumes"):
        subj = nii_path.stem.replace(".nii", "")
        try:
            vol = load_volume(nii_path)

            if vol.ndim == 4:
                vol = vol[..., 0]
            if vol.ndim != 3:
                print(f"  [SKIP] {subj}: unexpected shape {vol.shape}")
                continue

            slice_indices = centre_slice_indices(vol.shape[2], MAX_SLICES)

            for z in slice_indices:
                slc = normalise(vol[:, :, z])
                if slc.max() < 0.01:
                    continue

                filtered = apply_filters(slc)
                mets = {fn: compute_metrics(slc, filtered[fn])
                        for fn in FILTER_NAMES}

                for fn in FILTER_NAMES:
                    p_val, s_val = mets[fn]
                    row = {"subject": subj, "slice": z,
                           "filter": fn, "psnr": p_val, "ssim": s_val}
                    writer.writerow(row)
                    csv_rows.append(row)

                if (z % SAVE_EVERY_N) == 0:
                    tag = f"{subj}_z{z:04d}"
                    save_png(slc,
                             OUT_DIR / "before" / f"{tag}_original.png")
                    for fn in FILTER_NAMES:
                        save_png(filtered[fn],
                                 OUT_DIR / "after" / fn / f"{tag}_{fn}.png")
                    save_comparison(
                        slc, filtered, mets,
                        OUT_DIR / "comparisons" / f"{tag}_comparison.png"
                    )

            volumes_processed += 1
            del vol
            gc.collect()

        except Exception as e:
            print(f"  [ERROR] {subj}: {e} â€” skipping")
            continue

print(f"\n{'='*50}")
print(f"Total volumes processed : {volumes_processed}")
print(f"Total CSV rows written  : {len(csv_rows)}")
print(f"\nMean metrics across all slices:")
print(f"  {'Filter':<12} {'Mean PSNR':>10} {'Mean SSIM':>10}")
print(f"  {'-'*34}")
for fn in FILTER_NAMES:
    rows = [r for r in csv_rows if r["filter"] == fn]
    if rows:
        mp = np.mean([r["psnr"] for r in rows])
        ms = np.mean([r["ssim"] for r in rows])
        print(f"  {fn.capitalize():<12} {mp:>10.4f} {ms:>10.4f}")
    else:
        print(f"  {fn.capitalize():<12} {'N/A':>10} {'N/A':>10}")

print(f"\nOutput saved to: {OUT_DIR.resolve()}")
print(f"  {OUT_DIR}/before/")
print(f"  {OUT_DIR}/after/gaussian/")
print(f"  {OUT_DIR}/after/mean/")
print(f"  {OUT_DIR}/after/median/")
print(f"  {OUT_DIR}/comparisons/")
print(f"  {OUT_DIR}/metrics.csv")
print(f"{'='*50}")
```
