# IMPORTS

```python
import os, gc, csv, warnings
import numpy as np
import nibabel as nib
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from pathlib import Path
from scipy.ndimage import (binary_erosion, binary_dilation,
                            binary_opening, binary_closing,
                            label as nd_label)
from skimage.feature import canny, graycomatrix, graycoprops
from skimage.filters import sobel
from skimage.morphology import disk
from skimage.measure import regionprops, label as sk_label
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import (classification_report, confusion_matrix,
                              accuracy_score, ConfusionMatrixDisplay)
from tqdm import tqdm
import kagglehub

warnings.filterwarnings("ignore")
```

# CONFIG

```python
MAX_VOLUMES  = 10
MAX_SLICES   = 5
CANNY_SIGMA  = 1.5
MORPH_ITER   = 2
GLCM_DISTS   = [1, 2]
GLCM_ANGLES  = [0, np.pi/4, np.pi/2, 3*np.pi/4]
OUT_DIR      = Path("output_a3")

for d in [OUT_DIR / "glcm_plots",
          OUT_DIR / "geometry_plots",
          OUT_DIR / "classification"]:
    d.mkdir(parents=True, exist_ok=True)
```

# DOWNLOAD

```python
print("Downloading dataset via kagglehub...")
path = kagglehub.dataset_download("awansaad6797/cancer-dataset")
print(f"Dataset path: {path}")

all_nii = sorted(f for f in Path(path).rglob("*.nii") if f.is_file())
all_nii = [f for f in all_nii if "seg" not in f.name.lower()]
all_nii = all_nii[:MAX_VOLUMES]
print(f"Using {len(all_nii)} volumes.")
```

# HELPERS

```python
def normalise(slc):
    mn, mx = slc.min(), slc.max()
    if mx > mn:
        return ((slc - mn) / (mx - mn + 1e-8)).astype(np.float32)
    return np.zeros_like(slc, dtype=np.float32)

def centre_slice_indices(n_slices, max_slices):
    if n_slices <= max_slices:
        return list(range(n_slices))
    start = (n_slices - max_slices) // 2
    return list(range(start, start + max_slices))

def get_clean_mask(slc):
    edges  = canny(slc, sigma=CANNY_SIGMA)
    closed = binary_closing(edges, structure=disk(4))
    labeled, n = nd_label(closed)
    if n == 0:
        return edges.astype(np.uint8)
    sizes   = [(labeled == i).sum() for i in range(1, n + 1)]
    mask    = (labeled == (np.argmax(sizes) + 1)).astype(np.uint8)
    struct  = disk(3).astype(bool)
    cleaned = binary_closing(mask, structure=struct, iterations=MORPH_ITER).astype(np.uint8)
    return cleaned
```

# TASK 1: GLCM

```python
def compute_glcm_features(slc, mask):
    region = slc.copy()
    region[mask == 0] = 0
    img_uint = (region * 63).astype(np.uint8)
    glcm = graycomatrix(img_uint, distances=GLCM_DISTS, angles=GLCM_ANGLES,
                        levels=64, symmetric=True, normed=True)
    energy      = graycoprops(glcm, "energy").mean()
    contrast    = graycoprops(glcm, "contrast").mean()
    homogeneity = graycoprops(glcm, "homogeneity").mean()
    correlation = graycoprops(glcm, "correlation").mean()
    glcm_norm   = glcm[:, :, 0, 0]
    entropy     = float(-np.sum(glcm_norm * np.log2(glcm_norm + 1e-10)))
    return {
        "energy":      round(float(energy),      6),
        "entropy":     round(entropy,             6),
        "contrast":    round(float(contrast),     6),
        "homogeneity": round(float(homogeneity),  6),
        "correlation": round(float(correlation),  6),
    }

def save_glcm_plot(slc, mask, features, tag):
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))
    axes[0].imshow(slc, cmap="gray"); axes[0].set_title("Original", fontsize=9); axes[0].axis("off")
    axes[1].imshow(slc, cmap="gray"); axes[1].imshow(mask, cmap="Reds", alpha=0.5)
    axes[1].set_title("Masked Region", fontsize=9); axes[1].axis("off")
    colours = ["#4C72B0", "#DD8452", "#55A868", "#C44E52", "#8172B2"]
    axes[2].bar(features.keys(), features.values(), color=colours)
    axes[2].set_title("GLCM Features", fontsize=9)
    axes[2].tick_params(axis="x", rotation=25, labelsize=7)
    for i, v in enumerate(features.values()):
        axes[2].text(i, v + 0.001, f"{v:.4f}", ha="center", fontsize=6)
    fig.suptitle(f"GLCM â€“ {tag}", fontsize=11); fig.tight_layout()
    fig.savefig(OUT_DIR / "glcm_plots" / f"{tag}_glcm.png", dpi=90, bbox_inches="tight")
    plt.close(fig)
```

# TASK 2: GEOMETRIC FEATURES

```python
def compute_geometric_features(mask):
    labeled_mask = sk_label(mask)
    props = regionprops(labeled_mask)
    if not props:
        return {"area": 0, "centroid_r": 0, "centroid_c": 0, "perimeter": 0, "circularity": 0}
    region      = max(props, key=lambda r: r.area)
    area        = region.area
    centroid_r  = round(region.centroid[0], 2)
    centroid_c  = round(region.centroid[1], 2)
    perimeter   = round(region.perimeter, 4)
    circularity = round((4 * np.pi * area) / (perimeter ** 2 + 1e-8), 6)
    return {"area": area, "centroid_r": centroid_r, "centroid_c": centroid_c,
            "perimeter": perimeter, "circularity": circularity}

def save_geometry_plot(slc, mask, geo, tag):
    fig, axes = plt.subplots(1, 2, figsize=(10, 4))
    axes[0].imshow(slc, cmap="gray"); axes[0].imshow(mask, cmap="Blues", alpha=0.4)
    axes[0].plot(geo["centroid_c"], geo["centroid_r"], "r+", markersize=14, markeredgewidth=2)
    axes[0].set_title(f"Area={geo['area']}pxÂ²  Circ={geo['circularity']:.4f}", fontsize=8)
    axes[0].axis("off")
    geo_display = {"Area (pxÂ²)": geo["area"], "Perimeter": geo["perimeter"], "Circularity": geo["circularity"]}
    axes[1].bar(geo_display.keys(), geo_display.values(), color=["#2196F3", "#FF9800", "#4CAF50"])
    axes[1].set_title("Geometric Features", fontsize=9)
    axes[1].tick_params(axis="x", labelsize=8)
    for i, v in enumerate(geo_display.values()):
        axes[1].text(i, v * 1.01, f"{v:.2f}", ha="center", fontsize=7)
    fig.suptitle(f"Geometry â€“ {tag}", fontsize=11); fig.tight_layout()
    fig.savefig(OUT_DIR / "geometry_plots" / f"{tag}_geometry.png", dpi=90, bbox_inches="tight")
    plt.close(fig)
```

# MAIN PIPELINE

```python
all_features = []

for nii_path in tqdm(all_nii, desc="Volumes"):
    subj = nii_path.stem.replace(".nii", "")
    try:
        img = nib.load(str(nii_path))
        vol = img.get_fdata(dtype=np.float32)
        if vol.ndim == 4: vol = vol[..., 0]
        if vol.ndim != 3: continue

        for z in centre_slice_indices(vol.shape[2], MAX_SLICES):
            slc = normalise(vol[:, :, z])
            if slc.max() < 0.01: continue

            tag  = f"{subj}_z{z:04d}"
            mask = get_clean_mask(slc)
            if mask.sum() < 50: continue

            glcm_feats = compute_glcm_features(slc, mask)
            save_glcm_plot(slc, mask, glcm_feats, tag)

            geo_feats = compute_geometric_features(mask)
            save_geometry_plot(slc, mask, geo_feats, tag)

            fname = nii_path.name.lower()
            if   "t1c" in fname or "t1ce" in fname: label = "Enhancing"
            elif "t2f" in fname or "fl"   in fname: label = "Edema"
            elif "t1n" in fname or "t1"   in fname: label = "Necrotic"
            else:                                    label = "Other"

            row = {"subject": subj, "slice": z, "label": label, **glcm_feats, **geo_feats}
            all_features.append(row)

            print(f"  [{tag}] label={label}  energy={glcm_feats['energy']:.4f}  "
                  f"entropy={glcm_feats['entropy']:.4f}  area={geo_feats['area']}  "
                  f"circ={geo_feats['circularity']:.4f}")

        del vol; gc.collect()

    except Exception as e:
        print(f"  [ERROR] {subj}: {e} â€” skipping")

print(f"\nTotal feature vectors: {len(all_features)}")
```

# SAVE FEATURE VECTOR CSV

```python
csv_path = OUT_DIR / "feature_vectors.csv"
if all_features:
    with open(csv_path, "w", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=list(all_features[0].keys()))
        writer.writeheader(); writer.writerows(all_features)
    print(f"Feature vectors saved â†’ {csv_path}")
```

# TASK 3: CLASSIFICATION

```python
FEATURE_COLS = ["energy", "entropy", "contrast", "homogeneity",
                "correlation", "area", "perimeter", "circularity"]

if len(all_features) >= 10:
    X = np.array([[r[c] for c in FEATURE_COLS] for r in all_features])
    y = np.array([r["label"] for r in all_features])

    print(f"\nFeature matrix: {X.shape}")
    for lbl in np.unique(y):
        print(f"  {lbl}: {(y==lbl).sum()} samples")

    X_scaled = StandardScaler().fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=0.25, random_state=42)

    rf  = RandomForestClassifier(n_estimators=100, random_state=42)
    rf.fit(X_train, y_train); y_pred_rf = rf.predict(X_test)
    acc_rf = accuracy_score(y_test, y_pred_rf)

    svm = SVC(kernel="rbf", C=10, gamma="scale", random_state=42)
    svm.fit(X_train, y_train); y_pred_svm = svm.predict(X_test)
    acc_svm = accuracy_score(y_test, y_pred_svm)

    print(f"\n{'='*50}")
    print(f"  Random Forest Accuracy : {acc_rf*100:.2f}%")
    print(f"  SVM (RBF) Accuracy     : {acc_svm*100:.2f}%")
    print(f"{'='*50}")
    print("\nRandom Forest Report:"); print(classification_report(y_test, y_pred_rf, zero_division=0))
    print("SVM Report:");           print(classification_report(y_test, y_pred_svm, zero_division=0))

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    labels = sorted(np.unique(y))
    for ax, (title, y_pred) in zip(axes, [("Random Forest", y_pred_rf), ("SVM (RBF)", y_pred_svm)]):
        cm = confusion_matrix(y_test, y_pred, labels=labels)
        ConfusionMatrixDisplay(cm, display_labels=labels).plot(ax=ax, colorbar=False, cmap="Blues")
        ax.set_title(f"{title}  â€”  Accuracy={accuracy_score(y_test,y_pred)*100:.2f}%", fontsize=10)
        ax.tick_params(axis="x", rotation=30, labelsize=8)
    fig.suptitle("Confusion Matrices â€“ Tissue Classification", fontsize=12)
    fig.tight_layout()
    fig.savefig(OUT_DIR / "classification" / "confusion_matrices.png", dpi=100, bbox_inches="tight")
    plt.close(fig)

    importances = rf.feature_importances_
    sorted_idx  = np.argsort(importances)[::-1]
    fig, ax = plt.subplots(figsize=(9, 4))
    colours = ["#E53935" if i < 5 else "#1E88E5" for i in range(len(FEATURE_COLS))]
    ax.bar([FEATURE_COLS[i] for i in sorted_idx], importances[sorted_idx],
           color=[colours[i] for i in sorted_idx])
    ax.set_title("Random Forest â€“ Feature Importances", fontsize=11)
    ax.set_ylabel("Importance"); ax.tick_params(axis="x", rotation=30, labelsize=8)
    ax.legend(handles=[Patch(color="#E53935", label="GLCM"), Patch(color="#1E88E5", label="Geometric")], fontsize=8)
    fig.tight_layout()
    fig.savefig(OUT_DIR / "classification" / "feature_importance.png", dpi=100, bbox_inches="tight")
    plt.close(fig)

    with open(OUT_DIR / "classification" / "accuracy_report.csv", "w", newline="") as f:
        w = csv.writer(f)
        w.writerow(["Classifier", "Accuracy (%)", "Train Samples", "Test Samples"])
        w.writerow(["Random Forest", f"{acc_rf*100:.2f}", len(X_train), len(X_test)])
        w.writerow(["SVM (RBF)",     f"{acc_svm*100:.2f}", len(X_train), len(X_test)])
    print("Accuracy report saved.")

```
